```python
import numpy as np
import matplotlib.pyplot as plt
```

1a. Esboce o gráfico de x por f(x). Considerando Y = 4X − 3, esboce o gráfico de x por y.

```python
x = np.linspace(1, 4, 1000)
fx = np.ones_like(x) / 3  # Distribuição uniforme com f(x) = 1 / (b - a), onde a = 1 e b = 4

plt.plot(x, fx)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gráfico de x por f(x)')
plt.show()

y = 4 * x - 3

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Gráfico de x por y')
plt.show()
```

1b. Obtenha a função distribuição acumulada G(y).
Para calcular a função distribuição acumulada, podemos aplicar a transformação Y = 4X - 3 e resolver a equação para X:

Y = 4X - 3
X = (Y + 3) / 4

Agora, aplicamos a distribuição acumulada F(x) = (x - a) / (b - a) e substituímos X:

G(y) = F(X) = ((Y + 3) / 4 - 1) / (4 - 1)

```python
def G(y):
    return ((y + 3) / 4 - 1) / 3
```

1c. Obtenha a função densidade de probabilidade g(y). Esboce o gráfico de y por g(y).

Como a variável X é uniformemente distribuída, podemos encontrar a densidade de probabilidade de Y utilizando a derivada da função distribuição acumulada G(y) em relação a y:

g(y) = dG(y) / dy

```python
from sympy import symbols, diff

y = symbols('y')
G_sym = ((y + 3) / 4 - 1) / 3
gy_sym = diff(G_sym, y)

def g(y):
    return gy_sym.subs('y', y)

y_values = np.linspace(1, 13, 1000)
gy_values = [g(y_value) for y_value in y_values]

plt.plot(y_values, gy_values)
plt.xlabel('y')
plt.ylabel('g(y)')
plt.title('Gráfico de y por g(y)')
plt.show()
```

1d. Calcule P(Y > 10), indicando a área no gráfico esboçado no item anterior.

Aqui, utilizamos a função quad da biblioteca SciPy para integrar a função densidade de probabilidade g(y) no intervalo (10, 13) e calcular P(Y > 10). O valor dessa probabilidade e a área correspondente são mostrados no gráfico esboçado.

```python
from scipy.integrate import quad

result, _ = quad(g, 10, 13)

plt.fill_between(y_values[y_values > 10], 0, gy_values[y_values > 10], alpha=0.3)
plt.plot(y_values, gy_values)
plt.xlabel('y')
plt.ylabel('g(y)')
plt.title(f'Gráfico de y por g(y) com P(Y > 10) = {result:.4f}')
plt.show()
```
