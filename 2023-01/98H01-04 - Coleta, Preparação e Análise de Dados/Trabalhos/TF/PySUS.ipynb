```python
from pysus.preprocessing.sinan import read_sinan_dbf
from pysus.utilities.readdbc import read_dbc
from pysus.online_data import SIM
from pysus.online_data import CIHA
from pysus.online_data import SIA
from pysus.online_data import sinasc

import pandas as pd
import warnings

warnings.filterwarnings('ignore')
```

```python
#states = ['AC', 'AL', 'AP', 'AM', 'BA', 'CE', 'DF', 'ES', 'GO', 'MA', 'MT', 'MS', 'MG', 'PA', 'PB', 'PR', 'PE', 'PI', 'RJ', 'RN', 'RS', 'RO', 'RR', 'SC', 'SP', 'SE', 'TO']
states = ['MA', 'MT', 'MS', 'MG', 'PA', 'PB', 'PR', 'PE', 'PI', 'RJ', 'RN', 'RS', 'RO', 'RR', 'SC', 'SP', 'SE', 'TO']

years = list(range(1994, 2024))  # assumindo que você quer até o ano atual (2023)
```

```python
#SINASC data
def get_SINASC_data(state, year):
    sinascData = pd.DataFrame()
    try:
        sinascData = sinasc.download(state, year)
    
        sinascData['STATE'] = state
        sinascData['YEAR'] = year
    except:
        pass
    
    return sinascData
```

```python
###################################################SINASC DATA
```

```python
for state in states[:]:
    sinascData = pd.DataFrame()
    print(f'Loading SINASC - {state}')
    for year in years[:]:
        sinascData = sinascData.append(get_SINASC_data(state, year))
    
    minY = sinascData['YEAR'].min()
    maxY = sinascData['YEAR'].max()
    print(f'Saving SINASC - {state} {minY} to {maxY}')
    sinascData.to_parquet(f'SINASC_{state}_{minY}-{maxY}.parquet')
```
