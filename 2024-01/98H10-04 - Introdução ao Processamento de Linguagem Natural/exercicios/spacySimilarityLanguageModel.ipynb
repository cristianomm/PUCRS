```python
import spacy.cli
#from pathlib import Path
#spacy.cli.download("en_core_web_sm")
#spacy.cli.download("en_core_web_md")
spacy.cli.download("en_core_web_lg")

#Carregando o modelo e o texto
#nlp = spacy.load("en_core_web_sm")
nlp = spacy.load("en_core_web_lg")
texto = "A small building in the back. A small building in the brown."
doc = nlp(texto)
print(doc.text)
for sent in doc.sents:
  print(sent)
```

```python
#Similaridade baseada em modelo de linguagem
print(doc.text)
sentencas = list(doc.sents)
print("Similaridade entre as sentenças: ", sentencas[0].similarity(sentencas[1]))

for i in range(0,6):
  print("Similaridade entre tokens: ", sentencas[0][i], " com ", sentencas[1][i], ":", sentencas[0][i].similarity(sentencas[1][i]))

```
