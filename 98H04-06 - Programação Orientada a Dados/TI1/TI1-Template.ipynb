### Função auxiliar `mostra_dados`

````python
def mostra_dados(titulo, nome_arq, f):
    print("|___   ", titulo, "  ___|")
    with open(nome_arq, "r+", encoding='utf-8-sig') as arq:
        dados=csv.DictReader(arq, delimiter=',')
        fdados=f(dados)
        try:
            iterator=iter(fdados)
        except TypeError:
            print(fdados)
        else:
            for linha in fdados:
                print(linha)
````
