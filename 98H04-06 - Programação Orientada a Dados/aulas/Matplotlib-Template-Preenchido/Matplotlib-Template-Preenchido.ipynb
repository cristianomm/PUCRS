## Modos de Utilização

### Direto

```python
x = np.arange(1,6)
y = np.arange(1,6)

plt.plot(x,y)
plt.title('Meu Titulo')
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.show()
```

```python
plt.plot(np.random.randn(50).cumsum(), 'b--')
```

### Orientado a Objetos

```python
fig = plt.figure()

fig, ax = plt.subplots()

fig, ax = plt.subplots(2,2)

print(type(ax))
```

```python
x = np.arange(1,6)
y = np.arange(1,6)

fig, ax = plt.subplots()

ax.plot(x,y, label='Linear data')
ax.set_title('Meu Titulo')
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.legend()
plt.show()
```

## Gráfico de Linha ([`plot`](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.plot.html#matplotlib.axes.Axes.plot))

```python
x_dados = np.arange(40)
y_dados = np.random.randint(low=1, high=20, size=40)
z_dados = np.random.randint(low=1, high=30, size=40)
w_dados = np.random.randint(low=1, high=40, size=40)
```

```python
fig, ax = plt.subplots()
ax.plot(x_dados, y_dados, label='y_dados')
ax.plot(x_dados, z_dados, label='z_dados', color='b')
ax.plot(x_dados, w_dados, label='w_dados', color='#444444', linestyle='--', marker='o')


ax.set_title('Dados Aleatório')
ax.set_ylabel('Label y')
ax.set_xlabel('Label X')

ax.legend()

ax.grid()

ax.set_xticks(np.arange(0,40, step=2))

plt.savefig('plot.png')

print(plt.style.available)

#plt.style.use('Solarize_Light2')

plt.tight_layout()

plt.show()
```
